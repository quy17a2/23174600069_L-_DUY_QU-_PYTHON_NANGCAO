```python
import csv
import numpy as np
```

```python


# 1. Đọc dữ liệu từ file CSV
file_path = r'D:\bài tập thực hành\lab2\diem_hoc_phan.csv'
data = []

with open(file_path, mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # Bỏ qua tiêu đề
    for row in reader:
        data.append([row[0], row[1]] + list(map(float, row[2:])))

# Chuyển đổi dữ liệu thành mảng NumPy
data_np = np.array(data, dtype=object)

# Lấy điểm số (từ cột HP1, HP2, HP3)
hp_scores = data_np[:, 2:].astype(float)

# 2. Quy đổi điểm từ thang 10 sang điểm tín chỉ
def convert_to_credit(score):
    if score >= 8.5:
        return 'A'
    elif score >= 8.0:
        return 'B+'
    elif score >= 7.0:
        return 'B'
    elif score >= 6.5:
        return 'C+'
    elif score >= 5.5:
        return 'C'
    elif score >= 5.0:
        return 'D+'
    elif score >= 4.0:
        return 'D'
    else:
        return 'F'

# Áp dụng quy đổi
credit_grades = np.vectorize(convert_to_credit)(hp_scores)

# 3. Chia tách dữ liệu theo từng học phần
hp1_scores = hp_scores[:, 0]
hp2_scores = hp_scores[:, 1]
hp3_scores = hp_scores[:, 2]

# 4. Phân tích dữ liệu theo từng học phần
def analyze_scores(scores, hp_name):
    total = np.sum(scores)
    mean = np.mean(scores)
    std_dev = np.std(scores)
    return {
        "Học phần": hp_name,
        "Tổng điểm": total,
        "Điểm trung bình": mean,
        "Độ lệch chuẩn": std_dev
    }

hp1_analysis = analyze_scores(hp1_scores, "HP 1")
hp2_analysis = analyze_scores(hp2_scores, "HP 2")
hp3_analysis = analyze_scores(hp3_scores, "HP 3")

# 5. Phân tích điểm tổng hợp (tính số lượng từng loại điểm chữ trên tất cả học phần)
all_grades = credit_grades.flatten()  # Chuyển tất cả điểm chữ thành một mảng
unique_grades, grade_counts = np.unique(all_grades, return_counts=True)

# 6. Hiển thị danh sách từng sinh viên với điểm tín chỉ
def print_student_grades(data_np, credit_grades):
    print("\nDanh sách từng sinh viên và điểm tín chỉ:")
    print(f"{'ID':<5}{'Tên sinh viên':<20}{'HP 1':<10}{'HP 2':<10}{'HP 3':<10}{'HP 1 (Grade)':<15}{'HP 2 (Grade)':<15}{'HP 3 (Grade)':<15}")
    print("-" * 90)
    for i, row in enumerate(data_np):
        print(f"{row[0]:<5}{row[1]:<20}{row[2]:<10}{row[3]:<10}{row[4]:<10}{credit_grades[i][0]:<15}{credit_grades[i][1]:<15}{credit_grades[i][2]:<15}")

# Hiển thị kết quả phân tích
print("Phân tích từng học phần:")
print(hp1_analysis)
print(hp2_analysis)
print(hp3_analysis)

print("\nPhân tích điểm tổng hợp:")
for grade, count in zip(unique_grades, grade_counts):
    print(f"Điểm {grade}: {count} sinh viên")

# Gọi hàm để in danh sách
print_student_grades(data_np, credit_grades)
```
