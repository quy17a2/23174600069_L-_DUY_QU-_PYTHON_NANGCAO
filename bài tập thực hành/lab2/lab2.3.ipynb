```python
import numpy as np

# Hàm đọc dữ liệu từ tệp vào danh sách
def read_file_to_list(file_path):
    """
    Đọc dữ liệu từ tệp văn bản và lưu vào một danh sách.
    Mỗi dòng trong tệp được coi là một phần tử của danh sách.
    """
    with open(file_path, mode='r', encoding='utf-8') as file:
        return [line.strip() for line in file]

# Đường dẫn tới các tệp
efficiency_file = r'D:\bài tập thực hành\lab2\efficiency.txt'
shifts_file = r'D:\bài tập thực hành\lab2\shifts.txt'

# Đọc dữ liệu vào danh sách
efficiency = read_file_to_list(efficiency_file)
shifts = read_file_to_list(shifts_file)

# Chuyển đổi danh sách shifts thành mảng NumPy
np_efficiency = np.array(list(map(int, efficiency)))  # Chuyển đổi efficiency sang số nguyên
np_shifts = np.array(shifts)  # shifts là chuỗi nên không cần chuyển đổi

# Kiểm tra kết quả
print("Danh sách hiệu suất (efficiency):", efficiency)
print("Danh sách ca làm việc (shifts):", shifts)

print("\nNumpy Array np_efficiency:")
print(np_efficiency)
print("Kiểu dữ liệu của np_efficiency:", type(np_efficiency))
print("Kiểu phần tử trong np_efficiency:", np_efficiency.dtype)

print("\nNumpy Array np_shifts:")
print(np_shifts)
print("Kiểu dữ liệu của np_shifts:", type(np_shifts))
print("Kiểu phần tử trong np_shifts:", np_shifts.dtype)
morning_efficiency = np_efficiency[np_shifts == 'Morning']  # Lọc các phần tử của np_efficiency dựa trên np_shifts

average_morning_efficiency = morning_efficiency.mean()

print("Hiệu suất của nhân viên làm ca 'Morning':", morning_efficiency)
print("Hiệu suất trung bình của ca 'Morning':", average_morning_efficiency)
non_morning_efficiency = np_efficiency[np_shifts != 'Morning']  # Lọc các phần tử của np_efficiency không thuộc 'Morning'

# Tính trung bình hiệu suất của các ca không phải 'Morning'
if non_morning_efficiency.size > 0:  # Kiểm tra có dữ liệu hợp lệ
    average_non_morning_efficiency = non_morning_efficiency.mean()
    print("Hiệu suất của nhân viên làm các ca khác 'Morning':", non_morning_efficiency)
    print("Hiệu suất trung bình của các ca khác 'Morning':", average_non_morning_efficiency)
else:
    print("Không có nhân viên nào làm việc trong các ca khác 'Morning'.")

structured_dtype = [('shift', 'U10'), ('efficiency', 'float')]
workers = np.array(list(zip(np_shifts, np_efficiency)), dtype=structured_dtype)
sorted_workers = np.sort(workers, order='efficiency')

highest_efficiency = sorted_workers[-1]
lowest_efficiency = sorted_workers[0]

# Kết quả
print("Mảng workers ban đầu:")
print(workers)

print("\nMảng workers sau khi sắp xếp theo 'efficiency':")
print(sorted_workers)

print("\nCa làm việc có hiệu suất cao nhất:")
print(f"Shift: {highest_efficiency['shift']}, Efficiency: {highest_efficiency['efficiency']}")

print("\nCa làm việc có hiệu suất thấp nhất:")
print(f"Shift: {lowest_efficiency['shift']}, Efficiency: {lowest_efficiency['efficiency']}")
```
